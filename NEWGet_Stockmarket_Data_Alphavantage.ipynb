## Hier werden alle Ticker heruntergeladen.

Das benötigen Sie nicht, wenn Sie Ticker händisch suchen, was für Ihr Projekt vermutlich sinnvoll ist!

```python
import requests
from bs4 import BeautifulSoup

def get_sp500_tickers():
    url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.find('table', {'class': 'wikitable'})

    tickers = []
    for row in table.find_all('tr')[1:]:
        cells = row.find_all('td')
        if len(cells) > 0:
            ticker = cells[0].text.strip()
            tickers.append(ticker)

    return tickers

sp500_tickers = get_sp500_tickers()
print(sp500_tickers)
```

```python
import pandas as pd
from alpha_vantage.timeseries import TimeSeries


# Replace 'YOUR_API_KEY' with your actual Alpha Vantage API key
api_key = 'IFYZD2MB07FXDGPM' 
#api_Gkey = 'HTKT59G612SJTBWN'
all_data = pd.DataFrame()

for ticker in namen: 
    ts = TimeSeries(key=api_key, output_format='pandas')
    data, _ = ts.get_monthly_adjusted(symbol=ticker)
    data['Ticker'] = ticker
    data.reset_index(inplace=True)
    all_data = pd.concat([all_data, data], ignore_index=True)
```

```python
all_data.rename(columns={'index': 'date'}, inplace=True)
pivot_data = all_data.pivot(index='date', columns='Ticker', values='5. adjusted close')
pivot_data
pivot_data.to_csv('Aktien3.csv', index=True)
```
