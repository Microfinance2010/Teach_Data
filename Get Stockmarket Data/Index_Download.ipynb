```python
import yfinance as yf
import pandas as pd

# Definiere die Ticker, die du herunterladen möchtest
tickers = {
    "S&P_500": "^GSPC",  # S&P 500 Index
    "2Y_Treasury": "^IRX",  # Proxy für kurzfristige Anleihen (2 Jahre)
    "10Y_Treasury": "^TNX",  # 10-jährige US-Staatsanleihe
    "30Y_Treasury": "^TYX"   # 30-jährige US-Staatsanleihe
}

start_date = "2001-01-01"
end_date = "2024-10-31"

# Funktion zum Bestimmen des letzten Geschäftstags eines Monats
def get_last_business_day_of_month(df):
    return df.resample('M').last()

# Erstelle leere DataFrames für jeden Ticker
df_sp500 = pd.DataFrame()
df_2y_treasury = pd.DataFrame()
df_10y_treasury = pd.DataFrame()
df_30y_treasury = pd.DataFrame()

# Herunterladen und Verarbeiten der Daten für alle Ticker
for label, ticker in tickers.items():
    print(f"Downloading data for {label} ({ticker})...")
    
    # Lade die Daten auf Tagesbasis
    data = yf.download(ticker, start=start_date, end=end_date, interval="1d")
    
    # Prüfe, ob die Daten erfolgreich heruntergeladen wurden
    if data.empty:
        print(f"No data found for {ticker}. Skipping.")
        continue
    
    # Extrahiere die Adjusted Close-Daten
    adjusted_close = data["Adj Close"]
    
    # Extrahiere den letzten Geschäftstag jedes Monats
    last_business_day_data = get_last_business_day_of_month(adjusted_close)
    
    # Sicherstellen, dass keine leeren Daten vorhanden sind
    if last_business_day_data.empty:
        print(f"No valid data for the last business day of the month for {ticker}. Skipping.")
        continue
    
    # Speichern der Ergebnisse in den entsprechenden DataFrames
    if label == "S&P_500":
        df_sp500 = last_business_day_data.to_frame(name="S&P_500")
    elif label == "2Y_Treasury":
        df_2y_treasury = last_business_day_data.to_frame(name="2Y_Treasury")
    elif label == "10Y_Treasury":
        df_10y_treasury = last_business_day_data.to_frame(name="10Y_Treasury")
    elif label == "30Y_Treasury":
        df_30y_treasury = last_business_day_data.to_frame(name="30Y_Treasury")

# Mergen der DataFrames, um den finalen DataFrame zu erstellen
final_df = df_sp500
final_df = final_df.merge(df_2y_treasury, left_index=True, right_index=True)
final_df = final_df.merge(df_10y_treasury, left_index=True, right_index=True)
final_df = final_df.merge(df_30y_treasury, left_index=True, right_index=True)

# Zeige das Ergebnis
print(final_df.head())

```
